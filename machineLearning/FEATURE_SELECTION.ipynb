# Introduction

### **What is COPD?**
Chronic obstructive pulmonary disease (COPD) is a chronic inflammatory lung disease that causes obstructed airflow from the lungs. It is the third leading cause of death worldwide, causing 3.23 million deaths in 2019. COPD isn’t curable, but it can get better by not smoking, avoiding air pollution and getting vaccines. It can be controlled with medicines, oxygen and pulmonary rehabilitation.
Signs and symptoms of COPD may include:

- Shortness of breath, especially during physical activities
- Wheezing
- Chest tightness
- A chronic cough that may produce mucus (sputum) that may be clear, white, yellow or greenish
- Frequent respiratory infections
- Lack of energy
- Unintended weight loss (in later stages)
- Swelling in ankles, feet or legs
***

## **COPD Staging Systems**

To assess the severity and prognosis of COPD, two commonly used staging systems are the GOLD (Global Initiative for Chronic Obstructive Lung Disease) grade and the BODE index.

### Gold Grade :

The GOLD staging system categorizes COPD severity into four grades based on the severity of airflow limitation, as measured by spirometry (forced expiratory volume in one second, FEV1, and the ratio of FEV1 to forced vital capacity, FEV1/FVC)(if less than 0.70):

- GOLD Grade 1 (Mild): FEV1 ≥ 80% predicted.
- GOLD Grade 2 (Moderate): 50% ≤ FEV1 < 80% predicted.
- GOLD Grade 3 (Severe): 30% ≤ FEV1 < 50% predicted.
- GOLD Grade 4 (Very Severe): FEV1 < 30% predicted or FEV1 < 50% predicted plus chronic respiratory failure.

### Bode Index :

The BODE index is a tool that is used by healthcare professionals to predict the mortality rate (death rate) from chronic obstructive pulmonary disease (COPD).

The letters in 'BODE' stand for:

- **B**ody Mass Index (BMI) is a calculation made by comparing height vs weight.
- **A**irway Obstruction: Airway obstruction is measured by evaluating FEV1 — the amount of air that can be forcefully exhaled in one second after a deep breath.
- **D**yspnea refers to the degree of breathlessness someone experiences while living with COPD.
- **E**xercise tolerance refers to how well someone does on a 6-minute walk test.

## **Dataset Description**

Important features in the dataset:
1. EXPECTORATION : Indicates whether the individual experiences cough.
2. Pallor: Unusual paleness of the skin, often indicative of reduced blood flow or oxygenation
3. Icterus: Yellowing of the skin and eyes due to elevated levels of bilirubin in the blood, typically associated with liver dysfunction
4. Cyanosis: Bluish discoloration of the skin or mucous membranes caused by inadequate oxygenation of the blood
5. Clubbing: Abnormal enlargement of the fingertips and toes, often associated with chronic hypoxemia
6. Pedal Edema: Swelling of the lower extremities (feet and ankles) due to fluid retention
7. Lymphadenopathy: Enlargement of lymph nodes, which may indicate infection, inflammation, or malignancy
8. Raised JVP (Jugular Venous Pressure): Increased pressure in the jugular vein, often indicative of heart failure or fluid overload
9. ESR (Erythrocyte Sedimentation Rate): A non-specific marker of inflammation. Elevated ESR may indicate ongoing inflammation
10. CRP (C-Reactive Protein): Another marker of inflammation. Increased CRP levels can be indicative of inflammation in the body
11. LDH (Lactate Dehydrogenase): An enzyme present in various body tissues. Elevated LDH levels may be associated with tissue damage or inflammation, which can occur in COPD
12. BIL (T) (Total Bilirubin): Elevated bilirubin levels may suggest liver dysfunction, which can be related to COPD in some cases.
13. ALB (Albumin): Low levels of albumin can indicate malnutrition
14. SGOT (Serum Glutamic Oxaloacetic Transaminase): Also known as AST, it is an enzyme found in the liver. Elevated levels may suggest liver dysfunction.
15. SGPT (Serum Glutamic Pyruvic Transaminase): Also known as ALT, another liver enzyme. Similar to SGOT, elevated levels may indicate liver dysfunction.
16. S ALP (Serum Alkaline Phosphatase): An enzyme found in the liver and bones. Elevated levels can be associated with liver or bone disorders.
17. CREAT (Creatinine): A marker of kidney function. In COPD, especially during exacerbations, kidney function may be compromised.
18. 6mwd (pred) (6-Minute Walk Distance - Predicted): Measures the distance a person can walk in 6 minutes. It is a functional test often used to assess exercise tolerance.
19. 6MWD (6-Minute Walk Distance): Measures the distance a person can walk in 6 minutes. It is a functional test used to assess exercise tolerance and overall physical function in COPD patients. A shorter distance may indicate reduced exercise capacity and overall functional status.
20. mMRC Grade (Modified Medical Research Council Dyspnea Scale): A scale used to assess the severity of breathlessness in COPD. It ranges from 0 to 4, with higher scores indicating more severe breathlessness.
21. Number of Exacerbations in Previous Year: Refers to the frequency of COPD exacerbations (acute worsening of symptoms) experienced by the individual in the year preceding the assessment. A higher number of exacerbations may indicate more severe disease and increased healthcare utilization.
22. FEV1(Forced Expiratory Volume in 1 second): Measures the amount of air a person can forcefully exhale in one second
23. FVC(Forced Vital Capacity): It measures the total amount of air exhaled forcefully and completely after a maximum inhalation after bronchodilator administration.
24. It calculates the ratio of the volume of air exhaled in the first second of a forced expiration to the total volume of air exhaled during the FVC test

```python
# Filling missing values in Columns which have a duration Column related to it
def fill_missing_values_based_on_duration(data, symptom_column, duration_column):
    data[symptom_column] = data.apply(lambda row: 'No' if row[duration_column] == 0 else 'Yes' if pd.isnull(row[symptom_column]) else row[symptom_column], axis=1)
    return data

symptom_duration_pairs = [
    ('SHORTNESS_OF_BREATH','SHORTNESS_OF_BREATH_DURATION'),
    ('COUGH', 'COUGH_DURATION'),
    ('EXPECTORATION', 'EXPECTORATION_DURATION'),
    ('CHEST_TIGTNESS', 'CHEST_TIGTNESS_DURATION'),
    ('FEVER', 'FEVER_DURATION'),
    ('CHEST_PAIN', 'CHEST_PAIN_DURATION')
]

for symptom, duration in symptom_duration_pairs:
    df = fill_missing_values_based_on_duration(df, symptom, duration)
```

```python
# If a person is non-smoker and filling smoking material with non-smoker only and smoking years as 0
def update_smoking_exposure(data):
    data.loc[data['TYPE_OF_SMOKER'] == 'non smoker', 'CIGARETTE/BIDI/GANJA/BIOMASS_FUEL_EXPOSURE_IN_PACK_YEARS'] = 'non smoker'
    data.loc[data['TYPE_OF_SMOKER'] == 'non smoker', 'NO_OF_YEARS'] = '0 years'
    return data

# Apply the function to update smoking exposure columns
df = update_smoking_exposure(df)
```

```python
# Filling NaN values in a Categorical Column with mode
df['CYANOSIS'] = df['CYANOSIS'].fillna(df['CYANOSIS'].mode().iloc[0])
```

```python
# Define function to calculate 6MWD_pred based on gender and using Age, Height and Weight
def calculate_6MWD_pred(row):
    if pd.isnull(row['6mwd_(pred)']):  # Check if 6MWD_pred is null
        if row['SEX'] == 'Male':
            return round(7.57 * row['HEIGHT_IN_CM'] - 5.02 * row['AGE'] - 1.76 * row['WEIGHT_IN_KGS'] - 309)
        else:
            return round(2.11 * row['HEIGHT_IN_CM'] - 2.29 * row['AGE'] - 5.78 * row['WEIGHT_IN_KGS'] + 667)
    else:
        return row['6mwd_(pred)']  # Keep the existing value if not null

# Apply the function to calculate 6MWD_pred only for rows where it's null
df['6mwd_(pred)'] = df.apply(calculate_6MWD_pred, axis=1)
```

```python
def treat_outliers_with_iqr(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column_name] = df[column_name].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
    return df

for column in numerical_columns:
    df = treat_outliers_with_iqr(df, column)
```

```python
import pandas as pd


def remove_correlated_features(data, threshold=0.8):
    corr_matrix = data.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    filtered_data = data.drop(to_drop, axis=1)
    return filtered_data


# Usage
filtered_data = remove_correlated_features(df_encoded_copy, threshold=0.8)
filtered_data.shape
```

```python
from sklearn.feature_selection import VarianceThreshold


def remove_low_variance_features(data, threshold=0.1):
    selector = VarianceThreshold(threshold)
    selector.fit(data)
    selected_features = data.columns[selector.get_support()]
    filtered_data = data[selected_features]
    return filtered_data


# Usage
filtered_data = remove_low_variance_features(df_encoded_copy, threshold=0.1)
filtered_data.shape
```

#### 4. Dimensionality Reduction

Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


def apply_pca(data, n_components=10):
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data)
    pca = PCA(n_components=n_components)
    pca.fit(scaled_data)
    pca_data = pca.transform(scaled_data)
    return pd.DataFrame(
        data=pca_data, columns=[f"PC{i+1}" for i in range(n_components)]
    )


# Usage
pca_data = apply_pca(df_encoded_copy, n_components=10)
```
