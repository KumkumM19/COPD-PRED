```python
def calculate_6MWD_pred(row):
    if pd.isnull(row["6mwd (pred)"]):  # Check if 6MWD_pred is null
        if row["SEX"] == "Male":
            return round(
                7.57 * row["HEIGHT_IN_CM"]
                - 5.02 * row["AGE"]
                - 1.76 * row["WEIGHT_IN_KGS"]
                - 309
            )
        else:
            return round(
                2.11 * row["HEIGHT_IN_CM"]
                - 2.29 * row["AGE"]
                - 5.78 * row["WEIGHT_IN_KGS"]
                + 667
            )
    else:
        return row["6mwd (pred)"]  # Keep the existing value if not null

# Apply the function to calculate 6MWD_pred only for rows where it's null
df["6mwd (pred)"] = df.apply(calculate_6MWD_pred, axis=1)
```

```python
def treat_outliers_with_iqr(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column_name] = df[column_name].apply(
        lambda x: (
            upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x)
        )
    )
    return df


for column in numerical_columns:
    df = treat_outliers_with_iqr(df, column)
```

```python
# Convert data to appropriate data types
X_train_gold_grade = np.array(X_train_gold_grade, dtype=np.float32)
X_test_gold_grade = np.array(X_test_gold_grade, dtype=np.float32)
y_train_gold_grade = np.array(y_train_gold_grade, dtype=np.float32)
y_test_gold_grade = np.array(y_test_gold_grade, dtype=np.float32)

# One-hot encode target variables
y_train_gold_grade_encoded = tf.keras.utils.to_categorical(y_train_gold_grade, num_classes=4)
y_test_gold_grade_encoded = tf.keras.utils.to_categorical(y_test_gold_grade, num_classes=4)

# Compute class weights
class_weights = class_weight.compute_class_weight(class_weight="balanced", classes=np.unique(y_train_gold_grade), y=y_train_gold_grade)
class_weights_dict = dict(enumerate(class_weights))

# Normalize input features
scaler = StandardScaler()
X_train_gold_grade = scaler.fit_transform(X_train_gold_grade)
X_test_gold_grade = scaler.transform(X_test_gold_grade)
```
